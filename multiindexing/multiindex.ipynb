MultiIndex Series and Dataframe:
Series in 1D object and Dataframe Object:

Why multiindexing??
we may get data of higher order so in order to view that data in a dataframe we should have proper knowledge of multiindexing.

```python
#multiindex series
import pandas as pd
a=pd.Series([1,2,3,4,5,6,7,8],index=pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]]))
a
```

```python
#MultiIndex from tuple
index_val=[('cse',2019),('cse',2020),('cse',2021),('ece',2019),('ece',2020),('ece',2021)]

b=pd.Series(range(6),index=pd.MultiIndex.from_tuples(index_val))
b
```

```python
b['cse']
```

```python
#unstack
b=b.unstack()#series to dataframe
b
```

```python
b.stack()#dataframe to series
```

Multiindexing dataframe:

```python
df=pd.DataFrame([[1,3],[2,4],[3,5],[4,6],[5,7],[6,8],[7,0],[8,10]],index=pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]]),columns=['students','package'])
df
```

```python
#multiindexing from columns perspective
multiindex=pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
col_index=pd.MultiIndex.from_product([['Kathmandu','Chitawan'],['Avg_package','students']])
a=[[1,3,4,5],[2,4,6,7],[3,5,8,7],[4,6,7,9],[5,7,3,5],[6,8,6,5],[6,4,7,0],[3,5,8,10]]
df=pd.DataFrame(a,columns=col_index,index=multiindex)
df
```

stacking and unstacking


```python
df.stack()
```

```python
df.unstack()
```

```python
df.unstack().unstack()
```

Long vs wide data:
(i.)Wide format:
--> Wide format is where we have a single row for each data point with multiple columns to hold the values of various attribute.
(ii.) Long format:
--> is where, for each datapoint we have as many rows as the number of attibutes and each row contains the value of particulat attribute for a given data point.

```python
#Ex: wide format
a=pd.DataFrame([[2,3,4],[3,4,5]],columns=['Height','Weight','class'],index=['Ram','Shyam'])
a
```

```python
#Ex: long format
b=pd.DataFrame([2,3,4,3,4,5],index=pd.MultiIndex.from_product([['Ram','Shyam'],['Height','Weight','class']]))
b
```

```python
##melt()--> to convert wide format data to long formatted data
a.melt()
```
