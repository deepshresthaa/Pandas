Pivot Table:
The pivot table takes simple columnwise data as input, and groups the entries into a two  dimensional table that provides a multidimensional summarization of the data.


```python
import pandas as pd
import numpy as np
import seaborn as sbs
```

```python
df=sbs.load_dataset('tips')
df
```

```python
df.groupby(['sex','smoker'])['total_bill'].mean().unstack()
```

```python
#doing same thing with pivot_table()
df.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
#using aggfunc() -> to specify what should pivot_table() calculate
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum')
```

```python
df.pivot_table(index='sex',columns='smoker',aggfunc={'total_bill':'mean',"tip":'sum','size':'sum'})
```

```python
#multidimensional:
df.pivot_table(index=['sex','smoker'],columns=['day','time'])
```

```python
#margins
df.pivot_table(index='sex',columns='smoker',values='total_bill',margins=True)
```
