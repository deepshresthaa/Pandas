# Time Series:
A time series is a dataset that tracks a sample over time. In particular, a time series allows one to see what factors influence certain variables from period to period. time series analysis can be useful to see how a given assets , security, or economic variable changes over time.
### Example:
- Financial Data(company stocks)
- Natural Data(Rainfall measurement)
- Event Data(Covid)
- Medical Data(Heart rate monitoring)

### Types of operations done on time series:
- Time series analysis
- Time series forecasting

# Resampling
Resampling involves changing the frequency of your time series observations.

Two types of resampling are:

Upsampling: Where you increase the frequency of the samples, such as from minutes to seconds.

Downsampling: Where you decrease the frequency of the samples, such as from days to months.

```python
#downsampling:
df['Close'].plot()
df['Close'].resample("Y").mean().plot()
```

```python
# df['Close'].plot()
df['Close'].resample('15D').mean().plot()
```

```python
#upsasampling: Usually upsampling is not done because noise will increase in the data
df["Close"].resample('12H').interpolate(method='spline',order=2).plot()
```

### Rolling Windows(Smoothing)
Time series data in original format can be quite volatile, especially on smaller aggregation levels. The concept of rolling, or moving averages is a useful technique for smoothing time series data.

```python
df['Close'].rolling(30).mean().plot()
df['Close'].ewm(30).mean().plot()
```
