```python
#strip()==> removes all leading and  trailing spaces
'          deep           '.strip()
#use case: user may enter extra spaces while entering name and address so inorder to handle it , we use strip()
```

```python
df['Name'].str.strip()
```

```python
#split() and get()

#ex: to extract salutation, lastname and firstname of each passenger and store it in new columns
df['lastname']=df['Name'].str.split(',').str.get(0)
```

```python
df[['title','firstname']]=df['Name'].str.split(', ').str.get(1).str.strip().str.split(" ",n=1,expand=True)
#Braund, Mr. Owen Harris
```

```python
df.sample()
```

```python
df['title'].value_counts()
```

```python
df['title']=df['title'].replace('Ms.','Miss.')
df['title']=df['title'].replace('Mlle.','Miss.')
```

```python
df.sample()
```

```python
df['title'].value_counts()
```

```python
#filtering
#ex:passenger name starting with d
df[df['firstname'].str.startswith('D')]
```

```python
#apply regular expression
#contains()
#ex: find name of passengers who has 'john' in his name
df[df['Name'].str.contains("john",case=False)]
```

```python
#'lastname' start and end with vowel character
df['lastname'][df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```
